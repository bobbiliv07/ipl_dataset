```python
def players_of_team(team_name):
    batters=data_set[data_set["batting_team"] == team_name]["batter"]
    non_striker=data_set[data_set["batting_team"] == team_name]["non_striker"]
    bowler=data_set[data_set["bowling_team"] == team_name]["bowler"]
    total_team_record=pd.concat([batters,bowler,non_striker])
    team_players=total_team_record.unique()
    bowlers=bowler.unique()
    batsmen=pd.concat([batters,non_striker]).unique()
    return team_players
```

```python
def player_played_for(player):
    batters=data_set[data_set["batter"] == player]["batting_team"]
    non_striker=data_set[data_set["non_striker"] == player]["batting_team"]
    bowler=data_set[data_set["bowler"] == player]["bowling_team"]
    all_teams=pd.concat([batters,bowler,non_striker])
    all_teams=all_teams.unique()
    return all_teams
```

```python
def get_top_run_scorers(data_set):
    top_scorers = data_set.groupby('batter',)['batsman_runs'].sum()
    print(top_scorers.sort_values(ascending=False).head(10))
```

```python
def get_top_wicket_takers(df):
    bowler_wickets = ['caught', 'bowled', 'lbw', 'stumped', 'caught and bowled', 'hit wicket']
    wickets_df = df[df['dismissal_kind'].isin(bowler_wickets)]
    top_bowlers = wickets_df.groupby('bowler').size()
    print(top_bowlers.sort_values(ascending=False).head(10))
```

```python
get_top_wicket_takers(data_set)
```

```python
def analyze_toss_impact(df):
    toss_win_match_win = df[df['toss_winner'] == df['winner']]
    percentage = (len(toss_win_match_win) / len(df)) * 100
    print(f"Percentage of matches won by the team that won the toss: {percentage:.2f}%")
    decision_impact = toss_win_match_win['toss_decision'].value_counts()
    print(decision_impact)
```

```python
def analyze_death_overs(df):
    death_overs_df = df[df['over'] >= 16]
    batsman_stats = death_overs_df.groupby('batter').agg(
        runs_scored=('batsman_runs', 'sum'),
        balls_faced=('ball', 'count')
    )
    # Filter for batsmen who have faced at least 100 balls in the death overs.
    batsman_stats = batsman_stats[batsman_stats['balls_faced'] >= 100]
    batsman_stats['strike_rate'] = (batsman_stats['runs_scored'] / batsman_stats['balls_faced']) * 100
    print("Top 5 Death Over Batsmen (min. 100 balls):")
    print(batsman_stats.sort_values(by='runs_scored', ascending=False).head(5))

    # Top 5 Death Over Bowlers (by Economy Rate)
    print("\nTop 5 Death Over Bowlers (min. 100 balls):")
    bowler_stats = death_overs_df.groupby('bowler').agg(
        runs_conceded=('total_runs', 'sum'),
        balls_bowled=('ball', 'count')
    )
    # Filter for bowlers who have bowled at least 100 balls.
    bowler_stats = bowler_stats[bowler_stats['balls_bowled'] >= 100]
    bowler_stats['economy_rate'] = bowler_stats['runs_conceded'] / (bowler_stats['balls_bowled'] / 6)
    print(bowler_stats.sort_values(by='economy_rate', ascending=True).head(5))
    print("-" * 60 + "\n")
    
```

```python
analyze_death_overs(data_set)
```
